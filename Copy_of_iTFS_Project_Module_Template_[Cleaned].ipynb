# 4. A gas turbine within a combined cycle

Gas turbines operate at extremely high temperatures and push the boundaries of material science and fluid mechanics. Most natural gas power plants operate using gas turbines and are regularly combined. Before they are deployed, they must be thoroughly tested for performance. Companies like General Electric and Siemens have a departments purely dedicated to such testing. Watch the following videos to get yourself familiar with gas turbine technology.

> [Testing Turbines](https://youtu.be/HewsOwCLi3I)

> [Combined Cycles](https://youtu.be/KVjtFXWe9Eo)

Gas turbines are modelled as air-standard Brayton Cycles. The cycle consists of a compressor, combustion chamber and a turbine. If a Brayton cycle were to be a continuous cycle it would nearly resemble a Rankine cycle with the working fluid being recycled throughout; and the combustion chamber being swapped for a more traditional heat exchanger.
